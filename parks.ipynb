# **Parks**

### **Project Goal**
Here I will look at the number of parks across the City of Pittsburgh. I will then create a data-driven argument to present the best neighborhood based on the highest number of parks.

### **Why Parks?**
Parks are beneficial to quality of life and provide a great environment for raising a family. Green spaces can help with mental health which improves quality of life. They're also great places for kids to play, engage with the community, and step away from the urban lifestyle when they so choose.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the data set and counting each time a neighborhood appears, the total count being the number of parks in the neighborhood.
park_df = pd.read_csv("parks.csv")
park_count = park_df['neighborhood'].value_counts()
#Displaying the top 10 neighborhoods by highest number of parks
park_count_df = park_count.reset_index()
park_count_df.columns = ['Neighborhood', 'Park_Count']
park_count_df.head(10)
```

```python
#Creating a bar graph with the top 10 neighborhoods in descending order
top_10_parks = park_count_df.head(10)
top_10_parks.plot.bar(x='Neighborhood', y='Park_Count', rot=65, title="Top 10 Neighborhoods by Parks")
```

# **Conclusion**
#### After analyzing the parks data set, I was able to create a list and graph displaying the 10 best neighborhoods using the metric that the more parks the better the neighborhood. 
#### It is clear that East Liberty is the best neighborhood with the highest number of parks at 12 parks. This is double the amount that neighborhoods 6-10 have. For a prospective resident who cares about the number of parks in a neighborhood, East Liberty is their best bet, but any of the top 10 neighborhoods should satisfy this desire.
